```python
import sys
sys.path.insert(0, "../"*2)
import warnings
warnings.filterwarnings('ignore')
```

```python
from quantumnetworks import SingleModeSystem, plot_full_evolution, Average, Std
import numpy as np
```

```python
A_in = lambda t: 0
sys = SingleModeSystem(params={"omega_a": 1.0, "kappa_a": 0.0, "kerr_a":0.0, "gamma_a":0.1}, A_in = A_in)
x_0 = np.array([1,0])
ts = np.linspace(0, 100, 10001)
X = sys.forward_euler(x_0, ts)
fig, ax = plot_full_evolution(X, ts, labels=["$q_a$","$p_a$"])
ax.legend()
```

```python
avg = Average(sys, x_0, ts)
avg.calculate(X)
```

```python
avg.sensitivity("omega_a")
```

```python
std = Std(sys, x_0, ts)
std.calculate(X)
```

```python
std.sensitivity("kappa_a")
```
