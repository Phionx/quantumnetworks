```python
from quantumnetworks import MultiModeSystem, plot_full_evolution
import numpy as np
```

```python
# params directly provided
omegas = [1,2]
kappas = [0.001,0.005]
gammas = [0.002,0.002]
kerrs = [0.001, 0.001]
couplings = [[0,1,0.002]]
sys = MultiModeSystem(params={"omegas":omegas, "kappas":kappas, "gammas":gammas, "kerrs": kerrs, "couplings":couplings})
x_0 = np.array([1,0,0,1])
ts = np.linspace(0, 10, 10001)
print(10/1000)
```

```python
X = sys.trapezoidal(x_0, ts)
```

```python
fig, ax = plot_full_evolution(X, ts, labels=["$q_a$","$p_a$","$q_b$","$p_b$"])
ax.set_xlim([0,6])
ax.legend()
```

```python
X_dynamic, ts_dynamic = sys.trapezoidal(x_0, ts, dynamic_dt=True)
```

```python
fig, ax = plot_full_evolution(X_dynamic, ts_dynamic, labels=["$q_a$","$p_a$","$q_b$","$p_b$"])
ax.set_xlim([0,6])
ax.legend()
```

Dynamic stepping is slower.
