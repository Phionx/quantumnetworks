```python
from quantumnetworks import ExpSystem, plot_evolution
import numpy as np
```

# Trapezoidal Method

```python
sys = ExpSystem(params={"lambda": 10})
x_0 = np.array([1])
ts = np.linspace(0, 1, 101)
X = sys.trapezoidal(x_0, ts)
plot_evolution(X[0,:], ts)
```

# Forward Euler

```python
sys = ExpSystem(params={"lambda": 10})
x_0 = np.array([1])
ts = np.linspace(0, 1, 101)
X = sys.forward_euler(x_0, ts)
plot_evolution(X[0,:], ts)
```

```python
u = sys.eval_u(0)
sys.eval_Jf(x_0, u)
```
